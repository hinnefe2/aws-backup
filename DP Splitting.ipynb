```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set(font='serif')
sns.set_style("dark")
sns.set_context('talk')
```

```python
# MODEL PARAMETERS
V_extrinsic = -0.5   # extrinsic doping from fabrication procedures

dip_offset = -1.0    # starting value of dipole effect (on/off changes by +1.0)
dip_switch = 0.5     # applied backgate that causes dipoles to switch (+/- symmetric)
dip_strength = 0.5   # size of dipole gating effect
dip_sharpness = 0.1  # sharpness of dipole transition
dip_step_up = True   # whether the dipole function is a step up or step down, viewed left to right



def smearedStepFn(x, x0, kT):
    return 1 - 1 / (np.exp((x-x0)/kT) + 1)

# carriers are considered in absolute number terms (electron-hole symmetry assumed)
# returns abs(V_eff) + 0.1 so that carriers doesn't go to 0
def smeared_carriers(V_eff):
    return abs(V_eff) + 0.1

# returns gating voltage due to dipole effect at given applied backgate voltage and sweep direction (sweep_up=boolean)
def V_dipole(V_bg, sweep_up):
    sign = 2*(-0.5 + int(dip_step_up)) # +1 for dip_step_up==True, -1 for False
    if sweep_up:
        return dip_strength * (dip_offset + smearedStepFn(sign*V_bg, sign*dip_switch, dip_sharpness))
    else:
        return dip_strength * (dip_offset + smearedStepFn(sign*V_bg, -sign*dip_switch, dip_sharpness))
    
# returns effective gating voltage in non-dipole region for given applied backgate voltage
def V_eff_nondip(V_bg):
    return V_bg + V_extrinsic

# returns effective gating voltage in dipole region for given applied backgate voltage
def V_eff_dip(V_bg, sweep_up):
    return V_eff_nondip(V_bg) + V_dipole(V_bg, sweep_up)

# returns the resistance at a given effective gate voltage
def resistance(V_bg, dip_fraction, sweep_up = True):
    # dip_fraction = fraction shielded by SiO2, ie fraction containg dipoles
    # from http://ecee.colorado.edu/~bart/book/mobility.htm
    q_e = 1 # 1.6E-19 coulombs
    geometry_factor = 1
    mu = 1000 # guess for mobility
    rho1 = 1 / (q_e * mu * smeared_carriers(V_eff_nondip(V_bg))) # assuming electron-hole symmetry
    rho2 = 1 / (q_e * mu * smeared_carriers(V_eff_dip(V_bg,sweep_up))) # assuming electron-hole symmetry
    R1 = rho1 * geometry_factor
    R2 = rho2 * geometry_factor
    return dip_fraction*R2 + (1-dip_fraction)*R1

def current(V_bg, dip_fraction, sweep_up):
    V_sd = 5.0E-3
    return V_sd / resistance(V_bg, dip_fraction, sweep_up)

V_bg_up = np.linspace(-1.0,1.5,1000)
V_bg_down = np.linspace(1.5,-1.0,1000)
```

```python
colors = sns.color_palette()
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,5), sharex=True);

ax1.set_title('I_sd - Sweep Up')
ax1.set_xlabel('V_gate applied (V)')
ax1.set_xlim([0.25,0.75]);
ax1.set_ylim([0.25,2]);
ax1.plot(V_bg_up, [current(v, 0.10, True) for v in V_bg_up],color=colors[0], label='0.10% shielded');
ax1.plot(V_bg_up, [current(v, 0.25, True) for v in V_bg_up],color=colors[1], label='0.25% shielded');
ax1.plot(V_bg_up, [current(v, 0.50, True) for v in V_bg_up],color=colors[2], label='0.50% shielded');
ax1.plot(V_bg_up, [current(v, 0.75, True) for v in V_bg_up],color=colors[3], label='0.75% shielded');
ax1.legend(loc=3,fancybox=True, framealpha=0.1);

ax2.set_title('I_sd - Sweep Down')
ax2.set_xlabel('V_gate applied (V)')
ax2.set_xlim([0.25,0.75]);
ax2.set_ylim([0.25,2]);
ax2.plot(V_bg_down, [current(v, 0.10, False) for v in V_bg_down],color=colors[0], label='0.10% shielded');
ax2.plot(V_bg_down, [current(v, 0.25, False) for v in V_bg_down],color=colors[1], label='0.25% shielded');
ax2.plot(V_bg_down, [current(v, 0.50, False) for v in V_bg_down],color=colors[2], label='0.50% shielded');
ax2.plot(V_bg_down, [current(v, 0.75, False) for v in V_bg_down],color=colors[3], label='0.75% shielded');
ax2.legend(loc=3,fancybox=True, framealpha=0.9);
```

```python
colors = sns.color_palette()
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,5), sharex=True);

ax1.set_xlim([-1.1,1.6]);
ax1.set_title('Sweep Up')
ax1.set_xlabel('Applied V_gate (V)')
ax1.set_ylabel('Component contribution')
ax1.plot(V_bg_up, [0]*1000, '--', color=sns.xkcd_rgb['slate'], label='Dirac point')
ax1.plot(V_bg_up, V_bg_up, color=colors[0], alpha=0.5, label='applied backgate');
ax1.plot(V_bg_up, [V_extrinsic for v in V_bg_up], color=colors[1], alpha=0.5, label='extrinsic doping');
ax1.plot(V_bg_up, [V_dipole(v,True) for v in V_bg_up], color=colors[2], alpha=0.5, label='dipole contribution');
ax1.plot(V_bg_up, [V_eff_dip(v,True) for v in V_bg_up], color=sns.xkcd_rgb['grey'], linewidth=3, alpha=1, label='Shielded region total');
ax1.plot(V_bg_up, [V_eff_nondip(v) for v in V_bg_up], color=sns.xkcd_rgb['black'], linewidth=3, alpha=1, label='Non-shielded region total');
ax1.legend(loc=4);

ax2.set_title('Sweep Down')
ax2.set_xlabel('Applied V_gate (V)')
ax2.set_ylabel('Component contribution')
ax2.plot(V_bg_up, [0]*1000, '--', color=sns.xkcd_rgb['slate'], label='Dirac point')
ax2.plot(V_bg_down, V_bg_down, color=colors[0], alpha=0.5, label='applied backgate');
ax2.plot(V_bg_down, [V_dipole(v,False) for v in V_bg_down], color=colors[1], alpha=0.5, label='dipole contribution');
ax2.plot(V_bg_down, [V_extrinsic for v in V_bg_down], color=colors[2], alpha=0.5, label='extrinsic doping');
ax2.plot(V_bg_down, [V_eff_dip(v,False) for v in V_bg_down], color=sns.xkcd_rgb['grey'], linewidth=3, alpha=1, label='Shielded region total');
ax2.plot(V_bg_down, [V_eff_nondip(v) for v in V_bg_down], color=sns.xkcd_rgb['black'], linewidth=3, alpha=1, label='Non-shielded region total');
ax2.legend(loc=4);
```
