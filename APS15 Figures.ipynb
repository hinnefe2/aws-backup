```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

mpl.rcParams['lines.linewidth'] = 3
```

```python
V_bg_up = np.linspace(-1.0,1.0,1000)

def carriers(V_eff):
    return abs(V_eff) + 0.01

def V_eff(V_bg,offsets):
    return V_bg + sum(offsets)

def conductance(V_bg, dip_fraction, offsetsA, offsetsB):
    q_e = 1
    geometry_factor = 1
    mu = 1000
    rho1 = 1 / (q_e * mu * carriers(V_eff(V_bg, offsetsA)))
    rho2 = 1 / (q_e * mu * carriers(V_eff(V_bg, offsetsB)))
    R1 = rho1 * geometry_factor
    R2 = rho2 * geometry_factor
    return 1 / (dip_fraction*R2 + (1-dip_fraction)*R1)
```

```python
# plot 1

v_ext = 0.25
v_dip = -0.5

colors = ['b','g','r','black'] #sns.color_palette()
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(16,5), sharex=True);

ax1.set_xlim([-1.1,1.1]);
ax1.set_title('Total Gate Voltage')
ax1.set_xlabel('Applied V_gate (V)')
ax1.set_ylabel('Effective Gate Voltage')
ax1.plot(V_bg_up, [0]*1000, '--', color='grey', label='Dirac point')

ax2.set_title('Conductance vs Applied Gate Voltage')
ax2.set_xlabel('Applied V_gate (V)')
ax2.set_ylabel('Conductance')


# next two lines = just applied gate
ax1.plot(V_bg_up, V_bg_up, color=colors[0], alpha=1, label='applied backgate');
ax2.plot(V_bg_up, [conductance(v,0, [0],[0]) for v in V_bg_up],color=colors[0])

# next two lines = applied gate + extrinsic doping
#ax1.plot(V_bg_up, [v_ext]*1000, color=colors[1], alpha=1, label='extrinsic doping')
#ax2.plot(V_bg_up, [conductance(v,0, [v_ext],[0]) for v in V_bg_up],color=colors[1])

# next two lines = applied gate + extrinsic doping + dipole doping
#ax1.plot(V_bg_up, [v_dip]*1000, color=colors[2], alpha=1, label='extrinsic doping')
#ax2.plot(V_bg_up, [conductance(v,0, [v_ext + v_dip],[0]) for v in V_bg_up],color=colors[2])

ax1.legend(loc=2);

plt.savefig('Fig1-1.png', dpi=200, transparent=True)
```
