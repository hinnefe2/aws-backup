```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.optimize import curve_fit
```

```python
def get_data(filename, columns=None):
    
    # note that columns must be array like, for a single column place in brackets, like ['home_name']
    df = pd.read_csv(filename, sep=',', encoding='latin-1', skipinitialspace=True, index_col=1, usecols=columns)
    
    return df

# standard gaussian function
def gauss(x, *p):
    A, mu, sigma = p
    return A*np.exp(-(x-mu)**2/(2.*sigma**2))
```

```python
filenames = [str(yr)+'_hoops.csv' for yr in range(2002,2016)]
frames  = [get_data(f) for f in filenames]
data_full = pd.concat(frames, keys=filenames)
data_current = data_full.ix['2015_hoops.csv']
```

```python
# this cell: calculate home-court advantage from ppg and MOV

#name, away, home, diff
ppg_avgs = [] #= pd.DataFrame(columns=['year', 'team', 'away_ppg', 'home_ppg', 'diff_ppg'])
mov_avgs = [] #= pd.DataFrame(columns=['year', 'team', 'away_mov', 'home_mov', 'diff_mov'])

for f in filenames:
    year = f[:4]
    for team in team_list:
        away_games = data_full.ix[f]['away_name'] == team
        home_games = data_full.ix[f]['home_name'] == team
    
        away_ppg = data_full.ix[f]['away_final'][away_games].mean()
        home_ppg = data_full.ix[f]['home_final'][home_games].mean()
        diff_ppg = home_ppg - away_ppg
        s = {'year':year, 'team':team, 'away_ppg':away_ppg, 'home_ppg':home_ppg, 'diff_ppg':diff_ppg}
        ppg_avgs.append(s)
        
        away_mov = (data_full.ix[f]['away_final'][away_games] - data_full.ix[f]['home_final'][away_games]).mean()
        home_mov = (data_full.ix[f]['home_final'][home_games] - data_full.ix[f]['away_final'][home_games]).mean()
        diff_mov = home_mov - away_mov
        s = {'year':year, 'team':team, 'away_mov':away_mov, 'home_mov':home_mov, 'diff_mov':diff_mov}
        mov_avgs.append(s)
        
ppg = pd.DataFrame(ppg_avgs)
ppg = ppg[['year', 'team', 'away_ppg', 'home_ppg', 'diff_ppg']]
mov = pd.DataFrame(mov_avgs)
mov = mov[['year', 'team', 'away_mov', 'home_mov', 'diff_mov']]

#ppg_avgs.sort(key = lambda x: x[3]);
#mov_avgs.sort(key = lambda x: x[3]);
```

```python
# fit some gaussians
ppg_hist, ppg_bin_edges = np.histogram([c for c in zip(*ppg_avgs)][4], bins=range(-20,30,1), density=True);
ppg_bin_centres = (ppg_bin_edges[:-1] + ppg_bin_edges[1:])/2
mov_hist, mov_bin_edges = np.histogram([c for c in zip(*mov_avgs)][4], bins=range(-20,30,1), density=True);
mov_bin_centres = (mov_bin_edges[:-1] + mov_bin_edges[1:])/2
# p0 is the initial guess for the fitting coefficients (A, mu and sigma in gauss)
p0 = [1., 0., 1.]
ppg_coeff, ppg_var_matrix = curve_fit(gauss, ppg_bin_centres, ppg_hist, p0=p0)
mov_coeff, mov_var_matrix = curve_fit(gauss, mov_bin_centres, mov_hist, p0=p0)
# Get the fitted curve
ppg_hist_fit = gauss(ppg_bin_centres, *ppg_coeff)
mov_hist_fit = gauss(mov_bin_centres, *mov_coeff)

print('Fitted ppg mean = %.2f' % ppg_coeff[1])
print('Fitted ppg standard deviation = %.2f' % ppg_coeff[2])
print('Fitted mov mean = %.2f' % mov_coeff[1])
print('Fitted mov standard deviation = %.2f' % mov_coeff[2])

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(16,5));
ax1.set_xlim([-20,30]);
ax1.hist([c for c in zip(*ppg_avgs)][4], bins=range(-20,40,1));
ax1.plot(ppg_bin_centres, [4500*h for h in ppg_hist_fit], linewidth=3, label='Fitted data', color='black', alpha=0.5);
ax2.hist([c for c in zip(*mov_avgs)][4], bins=range(-20,40,1));
ax2.plot(mov_bin_centres, [4500*h for h in mov_hist_fit], linewidth=3, label='Fitted data', color='black', alpha=0.5);
```
